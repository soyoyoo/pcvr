```python

import os
import urllib

import numpy as np
gap_table = 'plx.google:ga_healthcheck_das.google_store_output.all'

sql = """
        select  fullVisitorId, ROUND(score, 4) as score, NTILE(5) OVER(ORDER BY score) as bucket
           FROM `"""+gap_table+"""`
 """

print ('# BigQuery SQL')  
print (sql)
    
# execute the query using datalab lib
import google.datalab.bigquery as bq
# Pandas lib to handle table data
import pandas as pd
query = bq.Query(sql)
query_result = query.execute()
query_data = query_result.result().to_dataframe()
    
print('# query_data # of users with score >= 0.5  {}'.format(query_data[query_data['score'] >= 0.5].shape[0]))
print(query_data[query_data['score']>=0.5].head(20))
print('# query_data # of users with score < 0.5 {}'.format(query_data[query_data['score'] < 0.5].shape[0]))
print(query_data[query_data['score'] < 0.5].head(20))
    
```

```python
query_data_with_index_column = query_data.reset_index();
schema = bq.Schema.from_data(query_data_with_index_column)
print schema
output_table = bq.Table('gap_bq_r.google_store_output')
output_table.create(schema,overwrite=True)
```
